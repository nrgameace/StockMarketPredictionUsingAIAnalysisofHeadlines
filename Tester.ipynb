```python
from xgboost import XGBClassifier
from sentence_transformers import SentenceTransformer
import torch
import numpy
import pandas as pd

#Load the trained model
model = XGBClassifier()
model.load_model('gpu_stock_model2.json')

userInput = input("Enter the article title you would like to convert: ")



#Setup for device and vector Model
device = 'cuda' if torch.cuda.is_available() else 'cpu'
vectorModel = SentenceTransformer('all-MiniLM-L6-v2', device=device)


#Generate the vector
embedding = vectorModel.encode(userInput)

#Create column names
columns = [f'vec_{i}' for i in range(len(embedding))]

#Wrap the embedding in a DataFrame to pass through
df_embedding = pd.DataFrame([embedding], columns=columns)

#Predict Value
y_pred = model.predict(df_embedding)
if y_pred[0] <= 0:
    print("Market Decreases")
else:
    print("Market Increases")
```
